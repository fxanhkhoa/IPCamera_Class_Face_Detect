```python
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('barca_face.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
## (Image, scalefactor, min_neighbor)
## scale factor -> Parameter specifying how much the image size is reduced at each image scale
## min neighbor -> Parameter specifying how many neighbors each candidate rectangle should have to retain it
## faces = face_cascade.detectMultiScale(gray, 1.7, 5) ## just got 2
faces = face_cascade.detectMultiScale(gray, 1.2, 3) ## got all
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

plt.imshow(img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
